```python
import os
import warnings
warnings.simplefilter('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from skimage.io import imread,imshow
from skimage.transform import resize
from skimage.color import rgb2gray
```

```python
Eva =os.listdir("C:/Users/sushm/Eva")
Oreo=os.listdir("C:/Users/sushm/Oreo")
Sush=os.listdir("C:/Users/sushm/Sush")
```

```python
limit=30
Eva_images=[None]*limit
j=0
for i in Eva:
    if(j<limit):
        Eva_images[j]=imread("C:/Users/sushm/Eva/"+i)
        j+=1
    else:
        break
        
```

```python
limit=30
Oreo_images=[None]*limit
j=0
for i in Oreo:
    if(j<limit):
        Oreo_images[j]=imread("C:/Users/sushm/Oreo/"+i)
        j+=1
    else:
        break
```

```python
limit=30
Sush_images=[None]*limit
j=0
for i in Sush:
    if(j<limit):
        Sush_images[j]=imread("C:/Users/sushm/Sush/"+i)
        j+=1
    else:
        break
```

```python
Eva_gray = [None]*limit
j=0
for i in Eva:
    if(j<limit):
        Eva_gray[j]=rgb2gray(Eva_images[j])
        j+=1
    else:
        break
```

```python
Oreo_gray = [None]*limit
j=0
for i in Oreo:
    if(j<limit):
        Oreo_gray[j]=rgb2gray(Oreo_images[j][:,:,0:3])
        j+=1
    else:
        break
```

```python
Sush_gray = [None]*limit
j=0
for i in Sush:
    if(j<limit):
        Sush_gray[j]=rgb2gray(Sush_images[j][:,:,0:3])
        j+=1
    else:
        break
```
